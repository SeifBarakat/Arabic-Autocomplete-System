```python
import pandas as pd
import re

df = pd.read_csv("/content/mnad_data/Le360.com.csv")

def clean_text(text):
    
    if not isinstance(text, str):
        return "" 

    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^\w\s\u0600-\u06FF]', ' ', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text 

cleaned_lines = []
for _, row in df.iterrows():
    title = str(row['Title']) if pd.notna(row['Title']) else ""
    body = str(row['Body']) if pd.notna(row['Body']) else ""
    full_text = f"{title} {body}"
    cleaned = clean_text(full_text)
    if len(cleaned) > 50:
        cleaned_lines.append(cleaned)

with open("mnad_cleaned.txt", "w", encoding="utf-8") as f:
    for line in cleaned_lines:
        f.write(line + "\n")
```

```python
from datasets import load_dataset

dataset = load_dataset("text", data_files={"train": "mnad_cleaned.txt"})
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "aubmindlab/aragpt2-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token
```

```python
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

tokenized_dataset = dataset.map(tokenize_function, batched=True)
```

```python
from transformers import Trainer, TrainingArguments, DataCollatorForLanguageModeling

training_args = TrainingArguments(
    output_dir="./results_mnad",
    overwrite_output_dir=True,
    num_train_epochs=1,
    per_device_train_batch_size=4,
    save_strategy="epoch",
    logging_dir="./logs_mnad",
    logging_steps=10,
    fp16=True,
    report_to="none",
    save_total_limit=1
)

data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    tokenizer=tokenizer,
    data_collator=data_collator
)
```

```python
trainer.train()
```

```python
trainer.save_model("./arabic_gpt2_finetuned_mnad")
tokenizer.save_pretrained("./arabic_gpt2_finetuned_mnad")
```

```python
import gradio as gr
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
import logging

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

model_path = "/content/drive/MyDrive/arabic_gpt2_finetuned_mnad"
try:
    model = AutoModelForCausalLM.from_pretrained(model_path)
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model.eval()
except Exception as e:
    logger.error(f"فشل في تحميل النموذج أو المحول: {e}")
    raise

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
logger.info(f"النموذج محمل على {device}")


def generate_completions(input_text, max_length, num_suggestions, temperature=0.7):
    if not input_text.strip():
        return ["يرجى إدخال نص لتوليد الاقتراحات."] * num_suggestions

    try:
        inputs = tokenizer(input_text, return_tensors="pt").to(device)
        with torch.no_grad():
            outputs = model.generate(
                inputs["input_ids"],
                max_length=int(max_length),
                num_return_sequences=int(num_suggestions),
                do_sample=True,
                top_p=0.9,
                temperature=temperature,
                pad_token_id=tokenizer.eos_token_id,
                no_repeat_ngram_size=2
            )

        completions = []
        for i in range(outputs.shape[0]):
            decoded = tokenizer.decode(outputs[i], skip_special_tokens=True).strip()
            if decoded not in completions and len(decoded) > len(input_text) + 10:
                completions.append(decoded)

        while len(completions) < num_suggestions:
            completions.append("لم يتم توليد المزيد من الاقتراحات.")

        return completions

    except Exception as e:
        logger.error(f"خطأ أثناء التوليد: {e}")
        return [f"حدث خطأ: {str(e)}"] * num_suggestions


def clear_inputs():
    return "", 50, 3, 0.7, "", "", ""


def create_gradio_interface():
    with gr.Blocks(theme=gr.themes.Soft()) as demo:
        gr.Markdown("""
        <h1 style="text-align: center; color: #1e3a8a">نظام الإكمال التلقائي للنصوص العربية</h1>
        <p style="text-align: center;">اكتب بداية الجملة بالعربية واحصل على اقتراحات ذكية.</p>
        """)

        with gr.Row():
            with gr.Column():
                input_text = gr.Textbox(label="النص المدخل", lines=3, rtl=True)
                max_length = gr.Slider(10, 150, value=50, step=5, label="أقصى طول للإكمال")
                num_suggestions = gr.Slider(1, 5, value=3, step=1, label="عدد الاقتراحات")
                temperature = gr.Slider(0.1, 1.0, value=0.7, step=0.1, label="درجة الإبداع")

                submit_button = gr.Button("إنشاء الاقتراحات")
                clear_button = gr.Button("مسح")

                gr.Examples([
                    ["اليوم هو يوم"],
                    ["في مدينة"],
                    ["كان هناك"]
                ], inputs=[input_text])

            with gr.Column():
                output1 = gr.Textbox(label="اقتراح 1", rtl=True, show_copy_button=True)
                output2 = gr.Textbox(label="اقتراح 2", rtl=True, show_copy_button=True)
                output3 = gr.Textbox(label="اقتراح 3", rtl=True, show_copy_button=True)

        submit_button.click(
            fn=generate_completions,
            inputs=[input_text, max_length, num_suggestions, temperature],
            outputs=[output1, output2, output3],
            show_progress=True
        )

        clear_button.click(
            fn=clear_inputs,
            outputs=[input_text, max_length, num_suggestions, temperature, output1, output2, output3]
        )

    return demo

if __name__ == "__main__":
    demo = create_gradio_interface()
    demo.launch()
```

```python
from datasets import Dataset
import pandas as pd
from sklearn.model_selection import train_test_split

with open("/content/mnad_cleaned.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

df = pd.DataFrame(lines, columns=["text"])

train_texts, test_texts = train_test_split(df, test_size=0.1, random_state=42)
test_dataset = Dataset.from_pandas(test_texts)

sample_size = 1000
test_texts_sample = test_texts.sample(n=sample_size, random_state=42).reset_index(drop=True)
test_dataset = Dataset.from_pandas(test_texts_sample)


def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

tokenized_test = test_dataset.map(tokenize_function, batched=True)

from transformers import Trainer, TrainingArguments, DataCollatorForLanguageModeling

training_args = TrainingArguments(
    output_dir="./eval_results",
    per_device_eval_batch_size=4,
    report_to="none"
)

data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

trainer = Trainer(
    model=model,
    args=training_args,
    tokenizer=tokenizer,
    data_collator=data_collator
)

eval_results = trainer.evaluate(eval_dataset=tokenized_test)

print("نتائج التقييم:")
for key, value in eval_results.items():
    print(f"{key}: {value}")

import math
perplexity = math.exp(eval_results["eval_loss"])
print(f"Perplexity: {perplexity}")
```
